```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Import dataset
wine = pd.read_csv('/content/wine.csv')
wine.head()
```

```python
wine.describe()
```

```python
wine = wine.iloc[:,1:]
wine
```

```python
wine.info
```

```python
wine_array = wine.values
wine_array
```

```python
#Normalize the data
from sklearn.preprocessing import scale
wine_norm = scale(wine_array)   
wine_norm
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(wine_norm)
scaled_data = scaler.transform(wine_norm)
```

# **PCA Implementation**

```python
#import PCA
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(scaled_data)
pca_values =  pca.transform(scaled_data)
pca_values

pca_values.shape
```

```python
plt.plot(pca_values)
plt.show
```

```python
#PCA components
pca.components_
```

```python
# The amount of variance that each PCA has
var = pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
Var = np.cumsum(np.round(var,decimals= 4)*100)
Var
```

```python
# Final Dataframe
final_df = pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Visualization of PCAs
fig = plt.figure(figsize=(12,8))
sns.scatterplot(data=final_df);
```

```python
sns.scatterplot(data=final_df, x='PC1', y='PC2', hue='Type')
```

```python
x = pca_values[:,0:1]
y = pca_values[:,1:2]
plt.scatter(x,y)
plt.show()
```

# **Check with other Clustering Algorithms**
**Hirarchical Clustering**

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# As we already have normalized data, create Dendrograms
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
# Create Clusters
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['Cluster_ID'])
y['Cluster_ID'].value_counts()
```

```python
# Adding clusters to dataset
wine2=wine.copy()
wine2['Cluster_ID']=hclusters.labels_
wine2
```

**K_Means**

```python
from sklearn.cluster import KMeans
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)

# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS');
```

**Use K=3**

```python
# Cluster algorithm using K=3
new_cluster=KMeans(3,random_state=30).fit(wine_norm)
new_cluster
```

```python
new_cluster.labels_
```

```python
# Assign clusters to the data set
wine2=wine.copy()
wine2['New_Cluster_ID']=new_cluster.labels_
wine2
```
